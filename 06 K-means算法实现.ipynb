导入依赖
make_blobs 用于产生本次实验所需的数据集，matplotlib 用于数据可视化，KMeans 用于 Kmeans 算法拟合构建

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

生成数据集

```python
X, y = make_blobs(n_samples=500, n_features=2, centers=4, random_state=1)
```

查看产生的数据集的维度信息

```python
print("X 的维度为：{}".format(X.shape))
print("y 的维度为：{}".format(y.shape))
```

绘制散点图

```python
fig, ax1 = plt.subplots(1)
ax1.scatter(X[:, 0], X[:, 1], marker="o", s=8)  # 设置点的形状为圆形  # 设置点的大小
plt.show()
```

根据生成数据集时对应的标签，按照不同颜色绘制散点图。

```python
color = ["red", "pink", "orange", "green"]
fig, ax1 = plt.subplots(1)

for i in range(4):
    ax1.scatter(X[y == i, 0], X[y == i, 1], marker="o", s=8, c=color[i])
    plt.show()
```

Kmeans 聚类

```python
n_clusters = 3
cluster1 = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
```

查看聚类后的每一个样本点的标签。

```python
y_pred1 = cluster1.labels_
print(y_pred1)
```

打印每个簇的质心。

```python
centroid1 = cluster1.cluster_centers_
print(centroid1)
```

可视化聚类结果

```python
color = ["red", "pink", "orange", "green"]

fig, ax1 = plt.subplots(1)

for i in range(n_clusters):
    ax1.scatter(X[y_pred1 == i, 0], X[y_pred1 == i, 1], marker="o", s=8, c=color[i])
    
ax1.scatter(centroid1[:, 0], centroid1[:, 1], marker="x", s=15, c="black")
plt.show()
```

再次 Kmeans 聚类

```python
n_clusters = 4
cluster2 = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
y_pred2 = cluster2.labels_
centroid2 = cluster2.cluster_centers_
print("质心：{}".format(centroid2))
```

可视化聚类结果

```python
color = ["red", "pink", "orange", "green"]

fig, ax1 = plt.subplots(1)

for i in range(n_clusters):
    ax1.scatter(X[y_pred2 == i, 0], X[y_pred2 == i, 1], marker="o", s=8, c=color[i])

ax1.scatter(centroid2[:, 0], centroid2[:, 1], marker="x", s=15, c="black")
plt.show()
```
